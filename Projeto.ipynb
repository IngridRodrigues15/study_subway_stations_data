### Estações mais antigas

```python
df = dados.sort_values(by="Idade", ascending=False).head(10)
df[['Nome', 'Idade']]
df1 = df
plt.figure(figsize=(10,8)) #(altura,largura)
plt.bar(df1.Nome.to_numpy(), df1.Idade.to_numpy(), color = 'Blue', label = 'Estações')
plt.xticks(rotation=75)
plt.title('ESTAÇÕES MAIS ANTIGAS DE SP', color = 'Blue')
plt.legend(loc="lower left")
plt.show()
```

QUANTIDADE, EM PERCENTAGEM, DE ESTAÇÕES PERTENCENTES A CADA EMPRESA DE METRÔ DA CIDADE DE SÃO PAULO

```python
dados['Construção'].value_counts().plot.pie(autopct="%.2f", figsize=(10,7))
plt.show()
```

Aqui podemos observar que o metro possui 48% das estações de São paulo enquanto a CPTM 51%, uma diferença minima mas facilmente esplicavel pelo fato que a CPTM é a unica a possuir estações nas cidades da grande são paulo.

GRÁFICO SEABORN SOBRE O NÚMERO DA LINHA E A SUA RESPECTIVA IDADE

### Quantidade de estações construidas por ano

```python
df1 = dados

df1['ano_inauguração'] = pd.to_datetime(df1['Inauguração']).dt.year

df1[['Nome', 'ano_inauguração']].groupby( 'ano_inauguração' ).count().plot.bar(figsize=(20,8), title= 'Quantidade de estações construidas por ano')
```

O ano que mais foram construidas  estações do metro foi no em 2018, sendo essas as estações com 1 ano de idade. 


Aqui podemos observar também, que os nos ultimos 20 anos, os que menos foram inauguradas estações foram 2001 e 2013, com apenas 1 estação cada um.
